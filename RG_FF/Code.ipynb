## Overall Final Output

```python
dataframe_total=dataframe_total[[
    'placekey', 'location_name', 'street_address', 'city', 'region','postal_code', 'latitude', 'longitude',
    'top_category', 'naics_code',
    'Month', 'spend_by_transaction_intermediary', 'online_spend','online_transactions', 
    'Week Start', 'Sum_spent_by_week','Average_spent_per_transaction_by_week', 
    'Month_FF','Raw_vistors_Monthly_level', 'Week Start_FF', 'visits_weekly_level']]
```

```python
#Set Postal Code to be int type
dataframe_total["postal_code"]=dataframe_total["postal_code"].apply(lambda x:int(x))
dataframe_total["postal_code"]=dataframe_total["postal_code"].astype(str)
#Naics_Code
for counter in range(0,len(dataframe_total["naics_code"])):
    naics_code=dataframe_total["naics_code"].iloc[counter]
    try:
        dataframe_total["naics_code"].iloc[counter]=int(naics_code)
    except:
        dataframe_total["naics_code"].iloc[counter]=np.nan
```

### a.Expand FF Raw Data

```python
dataframe_total_FF=dataframe_total[["placekey",'Month_FF','Raw_vistors_Monthly_level', 'Week Start_FF', 'visits_weekly_level']]
```

```python
#Week_Start_FF
dataframe_total_FF.loc[:,'Week Start_FF']=dataframe_total_FF['Week Start_FF'].apply(lambda x:x.strip('][').replace("'","").split(','))
dataframe_total_FF["Week Start_FF"]=dataframe_total_FF["Week Start_FF"].apply(lambda x:[temp.strip() for temp in x])
#Visits Weekly Level
dataframe_total_FF.loc[:,'visits_weekly_level']=dataframe_total_FF['visits_weekly_level'].apply(lambda x:x.strip('][').replace("'","").split(','))
dataframe_total_FF.loc[:,'visits_weekly_level']=dataframe_total_FF['visits_weekly_level'].apply(lambda x:[float(temp) for temp in x])
```

```python
#Month_FF
dataframe_total_FF.loc[:,'Month_FF']=dataframe_total_FF['Month_FF'].apply(lambda x:x.strip('][').replace("'","").split(','))
dataframe_total_FF["Month_FF"]=dataframe_total_FF["Month_FF"].apply(lambda x:[temp.strip() for temp in x])
#Raw Visitors Monthly Level
dataframe_total_FF.loc[:,'Raw_vistors_Monthly_level']=dataframe_total_FF['Raw_vistors_Monthly_level'].apply(lambda x:x.strip('][').replace("'","").split(','))
dataframe_total_FF.loc[:,'Raw_vistors_Monthly_level']=dataframe_total_FF['Raw_vistors_Monthly_level'].apply(lambda x:[float(temp) for temp in x])
```

### a.1 Weekly Level

```python
dataframe_total_FF_1=dataframe_total_FF[["placekey","Week Start_FF","visits_weekly_level"]]
dataframe_total_FF_1=dataframe_total_FF_1.explode(["Week Start_FF","visits_weekly_level"])
#Trim Week Start Date to get Month Date
dataframe_total_FF_1["Month_match"]=dataframe_total_FF_1["Week Start_FF"].apply(lambda x:x[0:7]+"-01")
#Set weekly visits to be integer 
dataframe_total_FF_1["visits_weekly_level"]=dataframe_total_FF_1["visits_weekly_level"].apply(lambda x:int(x))
dataframe_total_FF_1.head(2)
```

```python
dataframe_total_FF_1.shape[0]
```

### a.2 Monthly Level

```python
dataframe_total_FF_2=dataframe_total_FF[["placekey","Month_FF","Raw_vistors_Monthly_level"]]
dataframe_total_FF_2=dataframe_total_FF_2.explode(["Month_FF","Raw_vistors_Monthly_level"])
dataframe_total_FF_2.head(2)
```

### Merge FF_Raw_Data_Parts

```python
total_1=pd.merge(dataframe_total_FF_1,dataframe_total_FF_2,left_on=["placekey","Month_match"],right_on=["placekey","Month_FF"],how="outer")
total_1=total_1[["placekey","Month_match","Raw_vistors_Monthly_level","Week Start_FF","visits_weekly_level"]]
total_1.rename(columns={
    "Month_match":"Month_FF",
    "Raw_vistors_Monthly_level":"Raw_Visitors_Monthly_level",
    "visits_weekly_level":"Visits_Weekly_level"
},inplace=True)
#total_1["Raw_Visitors_Monthly_level"]=total_1["Raw_Visitors_Monthly_level"].fillna("Not Available")
total_1.head(2)
```

### Expand Revenues_and_Geometry

```python
total_2=dataframe_total[["placekey",'Month', 'spend_by_transaction_intermediary','online_spend', 'online_transactions', 
    'Week Start','Sum_spent_by_week', 'Average_spent_per_transaction_by_week']]
```

```python
#Week_Start
total_2.loc[:,'Week Start']=total_2['Week Start'].apply(lambda x:x.strip('][').replace("'","").split(','))
total_2["Week Start"]=total_2["Week Start"].apply(lambda x:[temp.strip() for temp in x])
#Visits Weekly Level
total_2.loc[:,'Sum_spent_by_week']=total_2['Sum_spent_by_week'].apply(lambda x:x.strip('][').replace("'","").split(','))
total_2.loc[:,'Sum_spent_by_week']=total_2['Sum_spent_by_week'].apply(lambda x:[float(temp) for temp in x])
total_2.loc[:,'Sum_spent_by_week']=total_2['Sum_spent_by_week'].apply(lambda x:[round(temp,2) for temp in x])
#Average Spent By week
total_2.loc[:,'Average_spent_per_transaction_by_week']=total_2['Average_spent_per_transaction_by_week'].apply(lambda x:x.strip('][').replace("'","").split(','))
total_2.loc[:,'Average_spent_per_transaction_by_week']=total_2['Average_spent_per_transaction_by_week'].apply(lambda x:[float(temp) for temp in x])
total_2.loc[:,'Average_spent_per_transaction_by_week']=total_2['Average_spent_per_transaction_by_week'].apply(lambda x:[round(temp,2) for temp in x])
```

```python
#Month
total_2.loc[:,'Month']=total_2['Month'].apply(lambda x:x.strip('][').replace("'","").split(','))
total_2["Month"]=total_2["Month"].apply(lambda x:[temp.strip() for temp in x])
```

### 1.a Weekly Level

```python
total_2_a=total_2[["placekey","Week Start","Sum_spent_by_week","Average_spent_per_transaction_by_week"]]
total_2_a=total_2_a.explode(["Week Start","Sum_spent_by_week","Average_spent_per_transaction_by_week"])
total_2_a.head()
```

### 2.b Monthly Level

```python
total_2_b=total_2[["placekey",'Month','online_spend', 'online_transactions']]
total_2_b.head()
```

```python
total_2_b_a=total_2_b[total_2_b["online_spend"]=="0"].explode(["Month"])
total_2_b_a.head()
```

```python
total_2_b_b=total_2_b[total_2_b["online_spend"]!="0"]
#Online Spend
total_2_b_b.loc[:,'online_spend']=total_2_b_b['online_spend'].apply(lambda x:x.strip('][').replace("'","").split(','))
total_2_b_b.loc[:,'online_spend']=total_2_b_b['online_spend'].apply(lambda x:[float(temp) for temp in x])
total_2_b_b.loc[:,'online_spend']=total_2_b_b['online_spend'].apply(lambda x:[round(temp,2) for temp in x])
#Online Transactions
total_2_b_b.loc[:,'online_transactions']=total_2_b_b['online_transactions'].apply(lambda x:x.strip('][').replace("'","").split(','))
total_2_b_b.loc[:,'online_transactions']=total_2_b_b['online_transactions'].apply(lambda x:[float(temp) for temp in x])
total_2_b_b.loc[:,'online_transactions']=total_2_b_b['online_transactions'].apply(lambda x:[round(temp,2) for temp in x])
total_2_b_b=total_2_b_b.explode(["Month","online_spend","online_transactions"])
total_2_b_b.head()
```

```python
total_2_b=pd.concat([total_2_b_a,total_2_b_b])
total_2_b.head()
```

### 2.c Spend Transaction Intermediary

```python
total_2_c=total_2[["placekey",'Month', 'spend_by_transaction_intermediary']]
total_2_c.head()
```

```python
total_2_c_a=total_2_c.loc[~total_2_c["spend_by_transaction_intermediary"].str.contains("]")].explode("Month")
total_2_c_a
```

```python
total_2_c_b=total_2_c.loc[total_2_c["spend_by_transaction_intermediary"].str.contains("]")]
total_2_c_b["spend_by_transaction_intermediary"]=total_2_c_b["spend_by_transaction_intermediary"].apply(lambda x:eval(x))
total_2_c_b=total_2_c_b.explode(["Month","spend_by_transaction_intermediary"])
```

```python
total_2_c=pd.concat([total_2_c_a,total_2_c_b])
total_2_c.head()
```

```python
#First Merge
total_2_2=pd.merge(total_2_b,total_2_c,on=["placekey","Month"])
total_2_2
```

```python
total_2_a["Month_Match"]=total_2_a["Week Start"].apply(lambda x:x[0:7]+"-01")
total_2_a
```

```python
total_2=pd.merge(total_2_a,total_2_2,left_on=["placekey","Month_Match"],right_on=["placekey","Month"],how="outer")
```

```python
total_2=total_2[["placekey","Month_Match","spend_by_transaction_intermediary","online_spend","online_transactions","Week Start","Sum_spent_by_week","Average_spent_per_transaction_by_week"]]
#total_3=total_3.fillna("Not Available")
```

```python
total_2.rename(columns={
    "Month_Match":"Month_RG",
    "Week Start":"Week Start_RG"
},inplace=True)
```

### Total_Output

```python
#Footfall Raw Data
total_1["Week Start_FF"]=total_1["Week Start_FF"].apply(lambda x:x.strip())
#Revenues and Geometry
total_2["Month_RG"]=total_2["Month_RG"].apply(lambda x:x.strip())
total_2["Week Start_RG"]=total_2["Week Start_RG"].apply(lambda x:x.strip())
```

```python
total_1_2=pd.merge(total_1,total_2,left_on=["placekey","Month_FF","Week Start_FF"],right_on=["placekey","Month_RG","Week Start_RG"],how="outer")
```

```python
total_1_2.loc[~total_1_2["spend_by_transaction_intermediary"].isna(),"spend_by_transaction_intermediary"]=total_1_2.loc[~total_1_2["spend_by_transaction_intermediary"].isna(),"spend_by_transaction_intermediary"].astype(str)
```
